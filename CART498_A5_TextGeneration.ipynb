```python
phrases_output = generator(
    phrase_prompt,

    max_new_tokens=60,    # ðŸ”¥ Prevents GPT-2 novel mode
    do_sample=True,
    temperature=1.0,
    top_k=50,
    num_return_sequences=10
)

phrases = []

for out in phrases_output:

    text = out["generated_text"]

    generated_part = text.split("Utterances:")[-1]

    lines = generated_part.split("\n")

    for line in lines:

        clean = line.strip()

        if 10 < len(clean) < 40:
            phrases.append(clean)

phrases[:16]
```

```python
name_prompt = """
Invent alien jungle creature designations.

Examples:
Kraval
Veshin
Torath
Neshar
Kalorin

Designations:
"""
```

```python
names_output = generator(
    name_prompt,

    max_new_tokens=40,     # ðŸ”¥ CRITICAL FIX
    do_sample=True,
    temperature=1.0,
    top_k=50,
    num_return_sequences=10
)

names = []

for out in names_output:
    text = out["generated_text"]

    generated_part = text.split("Designations:")[-1]

    for word in generated_part.split():
        clean = word.strip()

        if 4 < len(clean) < 12:
            names.append(clean)

names[:50]
```
